# Summer Analysis 

```python
import pandas as pd
```

```python
df = pd.read_csv("summer.csv")
```

# 1. In how many cities Summer Olympics is held so far?

```python
len(df['City'].unique())
```

# 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
r = df[df['Medal'] == 'Gold']
data = []

for sport in r['Sport'].unique():
    data.append([sport, len(r[r['Sport']  == sport])])
    
data = pd.DataFrame(data, columns = ['Sports','Gold Medals'])
data = data.sort_values(by = 'Gold Medals', ascending = False).head()
data.plot(x='Sports',y='Gold Medals', kind = 'bar')
print(data)
```
